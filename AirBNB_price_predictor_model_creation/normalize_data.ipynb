```python
lines = []
with open('AB_NYC_2019.csv','r',encoding='utf8') as file:
    for line in file:
        fields = line.strip().split(',')
        lines.append(fields)
lines
```

Eliminating records which may have length not equal to 13 because of additional commas for simplicity, as the number of such records id trivial.

```python
len([i for i in lines if len(i) > 13]) ##32
len([i for i in lines if len(i) < 13]) ##0
lines = [i for i in lines if len(i) == 13]
len(lines) ##48864
##removing records that may have ',' in host_name for simplification, as they have a low count and shouldnt affect the overall model
```

Picking out distinct records for the normalised 'NEIGHBOURHOOD' table

```python
neighbourhood_data = set()
for line in lines:
    nghbrgrp = line[3]
    nghbr = line[4]
    if nghbr != 'neighbourhood':
        neighbourhood_data.add((nghbr,nghbrgrp))
neighbourhood_data
```

Picking out distinct records for the normalised 'HOST' table

```python
host_data = set()
for line in lines:
    host_id = line[1]
    host_name = line[2]
    host_listings_count = line[11]
    if host_id != 'host_id':
        host_data.add((host_id,host_name,host_listings_count))
host_data
```

Adding neighbourhood_id for the primary key and creating the dictionary of neighbourhood_ids to neighbourhoods

```python
# neighbourhood_data = sorted(neighbourhood_data)
neighbourhood_data = [(i+1,neighbourhood_data[i][0],neighbourhood_data[i][1]) for i in range(0,len(neighbourhood_data))]
neighbourhood_data
```

```python
host_data = sorted([(int(host[0]),host[1],host[2]) for host in list(host_data)])
host_data
```

```python
neighbourhood_dict = {}
for ngh in neighbourhood_data:
    neighbourhood_dict[ngh[1]] = ngh[0]
neighbourhood_dict
```

Creating records for the main 'LISTINGS' table

```python
listings = []
for line in lines[1:]:
    try:
        id = int(line[0])
        hostid = int(line[1])
        nghbrid = neighbourhood_dict[line[4]]
        room_type = line[5]
        price = float(line[6])
        min_nights = int(line[7])
        nbr_reviews = int(line[8])
        reviews_per_month = float(line[10])
        availability_per_year = int(line[12])
        listings.append((id,hostid,nghbrid,room_type,price,min_nights,nbr_reviews,reviews_per_month,availability_per_year))
    except:
        continue
listings
```

Building the database

```python
import sqlite3
conn = sqlite3.connect('BNBListings.db')
conn.execute("PRAGMA foreign_keys = 1")
cur = conn.cursor()
cur.execute('''CREATE TABLE IF NOT EXISTS NEIGHBOURHOOD(
                NEIGHBOURHOOD_ID INTEGER PRIMARY KEY,
                NEIGHBOURHOOD TEXT,
                NEIGHBOURHOOD_GROUP TEXT)''')
cur.execute('''CREATE TABLE IF NOT EXISTS HOST(
                HOST_ID INTEGER PRIMARY KEY,
                HOST_NAME TEXT,
                NUMBER_OF_LISTINGS INTEGER)''')
cur.execute('''CREATE TABLE IF NOT EXISTS LISTINGS(
                ID INTEGER PRIMARY KEY,
                HOST_ID INTEGER,
                NEIGHBOURHOOD_ID INTEGER,
                ROOM_TYPE TEXT,
                PRICE NUMERIC,
                MIN_NIGHTS INTEGER,
                NUMBER_OF_REVIEWS INTEGER,
                REVIEWS_PER_MONTH NUMERIC,
                PER_YEAR_AVAILABILITY INTEGER,
                FOREIGN KEY (HOST_ID) REFERENCES HOST(HOST_ID),
                FOREIGN KEY (NEIGHBOURHOOD_ID) REFERENCES NEIGHBOURHOOD(NEIGHBOURHOOD_ID))''')
cur.executemany('INSERT INTO NEIGHBOURHOOD VALUES(?,?,?)',neighbourhood_data)
cur.executemany('INSERT INTO HOST VALUES(?,?,?)',host_data)
cur.executemany('INSERT INTO LISTINGS VALUES(?,?,?,?,?,?,?,?,?)',listings)
conn.commit()
conn.close()
```
